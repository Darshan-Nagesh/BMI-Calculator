```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
import matplotlib.pyplot as plt

# File path of the CSV dataset
csv_file_path = 'iris.csv'

# Read data from CSV file using pandas
data = pd.read_csv(csv_file_path)

# Extract the features from the dataset
features = data.iloc[:, :-1].values

# Perform agglomerative clustering using single linkage
linkage_matrix = linkage(features, method='single', metric='euclidean')
labels = fcluster(linkage_matrix, t=1.0, criterion='distance')

# Append cluster assignments to the dataset
data['Cluster'] = labels

# File path to write the clustered data to CSV file
clustered_csv_file_path = 'clustered_iris.csv'

# Write the clustered data to CSV file using pandas
data.to_csv(clustered_csv_file_path, index=False)

# Print the cluster assignments
print("Cluster assignments:")
print(labels)
print(data)
# Plot the dendrogram
plt.figure(figsize=(10, 6))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
dendrogram(linkage_matrix)
plt.show()
```

```python
# Perform agglomerative clustering using complete linkage
linkage_matrix = linkage(features, method='complete', metric='euclidean')
labels = fcluster(linkage_matrix, t=1.0, criterion='distance')

# Append cluster assignments to the dataset
data['Cluster'] = labels

# File path to write the clustered data to CSV file
clustered_csv_file_path = 'clustered_iris.csv'

# Write the clustered data to CSV file using pandas
data.to_csv(clustered_csv_file_path, index=False)

# Print the cluster assignments
print("Cluster assignments:")
print(labels)
print(data)
# Plot the dendrogram
plt.figure(figsize=(10, 6))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.grid(True)
dendrogram(linkage_matrix)
plt.show()
```
